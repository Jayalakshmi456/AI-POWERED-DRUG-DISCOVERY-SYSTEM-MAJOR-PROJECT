```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Load the dataset
data_path = '../data/raw/disease.csv'
df = pd.read_csv(data_path)

# Basic statistics
print("Dataset Shape:", df.shape)
print("\nSample Interactions:")
display(df.head())

# Distribution of interactions
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='interaction')
plt.title('Distribution of Drug-Protein Interactions')
plt.show()

# Sequence length distributions
df['protein_length'] = df['protein_sequence'].str.len()
df['drug_length'] = df['drug_sequence'].str.len()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
sns.histplot(data=df, x='protein_length', ax=ax1)
ax1.set_title('Protein Sequence Length Distribution')
sns.histplot(data=df, x='drug_length', ax=ax2)
ax2.set_title('Drug Sequence Length Distribution')
plt.show()
```
